merge the results from range and field of view (FOV)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Data
model_names = ['Natural', 'Accident', 'Homicide', 'Suicidal', 'Undetermined']

# Convert the data into dictionaries for easy plotting
lidar_range = [9.881965, 12.620561, 6.593451, 5.990318, 8.233798]
gaussian_range = [173.201345] * 5

lidar_fov = [171.525025, 245.181188, 83.023436, 70.949712, 126.062731]
gaussian_fov = [59997.412137] * 5

# Create the figure and axes
fig, ax1 = plt.subplots(figsize=(14, 8))

# Plotting the ranges
x = np.arange(len(model_names))
bar_width = 0.35

ax1.bar(x - bar_width/2, lidar_range, width=bar_width, color='skyblue', edgecolor='black', label='LiDAR Range')
ax1.bar(x + bar_width/2, gaussian_range, width=bar_width, color='lightcoral', edgecolor='black', label='Gaussian Range')

# Customize the first Y axis
ax1.set_xlabel('3D Models', fontsize=12)
ax1.set_ylabel('Range', fontsize=12)
ax1.set_title('Range and Field of View Comparison', fontsize=14)
ax1.set_xticks(x)
ax1.set_xticklabels(model_names, fontsize=10)
ax1.set_yscale('log')  # Log scale for better visibility
ax1.grid(True, which='both', linestyle='--', alpha=0.7, linewidth=0.7)
ax1.legend(loc='upper left', fontsize=10)

# Plotting Field of View (FOV) using a second Y axis
ax2 = ax1.twinx()
ax2.plot(x, lidar_fov, marker='o', linestyle='--', color='royalblue', linewidth=2, label='LiDAR FOV')
ax2.plot(x, gaussian_fov, marker='s', linestyle='--', color='indianred', linewidth=2, label='Gaussian FOV')

# Customize the second Y axis
ax2.set_ylabel('Field of View (FOV)', fontsize=12)
ax2.legend(loc='upper right', fontsize=10)

plt.tight_layout()
fig.savefig('Images/B3range_fov_comparison.png', dpi=200)
plt.show()
```
