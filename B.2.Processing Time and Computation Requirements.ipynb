To evaluate the computational resources required for each model, we can measure factors such as loading time, memory usage, and CPU/GPU usage. Below is a way to analyze the processing time and memory usage as a proxy for computational requirements, using Python's tracemalloc for memory tracking along with timing.

```python
import time
import trimesh
import tracemalloc

def load_model_and_measure(file_path, file_type):
    """Load a 3D model and measure time and memory."""
    tracemalloc.start()
    start_time = time.time()
    mesh = trimesh.load(file_path, file_type=file_type)
    end_time = time.time()
    current, peak = tracemalloc.get_traced_memory()
    tracemalloc.stop()
    processing_time = end_time - start_time
    peak_memory = peak / 10**6  # Convert to MB
    return processing_time, peak_memory

```

```python
# File paths for Nature
lidar_path_nature = '/Users/aaeshahalhakamy/MyResearch/Crime CV/Scenes/Nature/LiDARPLY/AdvancedExport/3DModel.ply'
gaussian_path_nature = '/Users/aaeshahalhakamy/MyResearch/Crime CV/Scenes/Nature/3DGS.ply'

# Measure computational requirements for Nature
lidar_time_nature, lidar_memory_nature = load_model_and_measure(lidar_path_nature, 'ply')
gaussian_time_nature, gaussian_memory_nature = load_model_and_measure(gaussian_path_nature, 'ply')

print("Nature - LiDAR:")
print(f"Time (s): {lidar_time_nature:.4f}, Memory (MB): {lidar_memory_nature:.2f}")

print("Nature - Gaussian Splatting:")
print(f"Time (s): {gaussian_time_nature:.4f}, Memory (MB): {gaussian_memory_nature:.2f}")
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
plt.rcParams['figure.dpi'] = 200
plt.rcParams['savefig.dpi'] = 200

# Aggregate data into a DataFrame
model_names = ['Natural', 'Accidental', 'Homicidal', 'Suicidal', 'Undetermined']
lidar_times = [lidar_time_nature, lidar_time_accident, lidar_time_homicide, lidar_time_suicide, lidar_time_undetermined]
gaussian_times = [gaussian_time_nature, gaussian_time_accident, gaussian_time_homicide, gaussian_time_suicide, gaussian_time_undetermined]
lidar_memory = [lidar_memory_nature, lidar_memory_accident, lidar_memory_homicide, lidar_memory_suicide, lidar_memory_undetermined]
gaussian_memory = [gaussian_memory_nature, gaussian_memory_accident, gaussian_memory_homicide, gaussian_memory_suicide, gaussian_memory_undetermined]

data = {
    'Model': model_names,
    'LiDAR Time (s)': lidar_times,
    'Gaussian Time (s)': gaussian_times,
    'LiDAR Memory (MB)': lidar_memory,
    'Gaussian Memory (MB)': gaussian_memory
}
df = pd.DataFrame(data)

# Print the DataFrame
print(df)

# Plotting the computational requirements
fig, ax = plt.subplots(figsize=(14, 8))

# Line plot for processing times
ax.plot(model_names, lidar_times, marker='o', linestyle='-', color='blue', label='LiDAR Time (s)', linewidth=2)
ax.plot(model_names, gaussian_times, marker='s', linestyle='--', color='orange', label='Gaussian Time (s)', linewidth=2)

# Line plot for memory usage
ax.plot(model_names, lidar_memory, marker='x', linestyle=':', color='green', label='LiDAR Memory (MB)', linewidth=2)
ax.plot(model_names, gaussian_memory, marker='^', linestyle='-.', color='red', label='Gaussian Memory (MB)', linewidth=2)

# Customize the plot
ax.set_xlabel('3D Models Scene', fontsize=14)
ax.set_ylabel('Processing Time and Computation Requirements', fontsize=14)
ax.set_title('Processing Time and Computation Requirements Comparison: Gaussian Splatting vs LiDAR', fontsize=16)
ax.set_xticks(np.arange(len(model_names)))
ax.set_xticklabels(model_names, fontsize=12)
ax.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)
ax.legend(fontsize=12)

plt.tight_layout()
plt.show()
fig.savefig('Images/B2.png', dpi=200)
```
